## Histogram, Probability Plot, Box Plot

Probability plots, by `stats.probplot` are useful for assessing whether a given data sample follows a particular theoretical distribution, such as the normal distribution. They can help answer questions like: "Does the data closely resemble a normal distribution?" or "Are there any significant deviations from the expected distribution?"

Here are a few reasons why probability plots, including the one generated by `stats.probplot`, are helpful:

- **Assessing distributional assumptions**: Probability plots provide a visual comparison between the observed data and the theoretical distribution. In the case of the `stats.probplot` function with `dist='norm'`, it compares the data against a normal distribution. By plotting the data on a standardized scale, deviations from the expected distribution become apparent. If the data points lie close to the straight line in the probability plot, it suggests that the data approximately follows a normal distribution.

- **Identifying departures from normality**: Probability plots can reveal departures from the expected distribution. If the data points deviate significantly from the straight line in the probability plot, it indicates that the data does not adhere closely to a normal distribution. This information can be valuable for selecting appropriate statistical methods or for deciding whether to apply transformations to achieve normality assumptions.

- **Detecting outliers**: These points can be seen as outliers if they fall far away from the straight line in the probability plot.

- **Validating model assumptions**: Probability plots are commonly used to validate assumptions made by statistical models. For example, linear regression assumes that the residuals (the differences between observed and predicted values) follow a normal distribution. By examining the probability plot of the residuals, we can assess whether this assumption holds. If the residuals deviate from the expected straight line, it suggests a violation of the normality assumption, prompting further investigation or model adjustments.

# Checking for outliers 

The upper and lower limits of a normal distribution can be determined using the standard deviation and mean of the distribution. In a standard normal distribution, where the mean is 0 and the standard deviation is 1, the common convention is to consider values within approximately 3 standard deviations from the mean as the "normal" range. 

To calculate the upper and lower limits for a normal distribution with a mean of 0 and a standard deviation of 1, we can multiply the standard deviation by 3 and add/subtract the result from the mean.

`Upper Limit: mean + (3 * standard deviation) = 0 + (3 * 1) = 3`

`Lower Limit: mean - (3 * standard deviation) = 0 - (3 * 1) = -3`

Therefore, for a standard normal distribution, the upper limit is 3 and the lower limit is -3. These limits represent approximately 99.7% of the data within the distribution, assuming it follows a perfect normal distribution.

```python
upperlimit = []
lowerlimit = []

for i in df.columns:
  upperlimit.append(df[i].mean()+(df[i].std())*4)
  lowerlimit.append(df[i].mean()-(df[i].std())*4)
```

**Many Features are co-related**

```python
def correlation(df, threshold):
  colcor = set()
  cormat = df.corr()

  for i in range(len(cormat)):
    for j in range(i):
      """
      for each cell get the value of that cell by iloc[i][j],
      if abs(value) > threshold, get the column name and add it in the set 
      """

      if abs(cormat.iloc[i][j])> threshold:
        colname = cormat.columns[i]
        colcor.add(colname)
  return colcor
```

```python
ans = correlation(train, threshold = 0.85)
ans
```

#Model Building 

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV 
from sklearn import metrics
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
X_train, X_test, y_train, y_test = train_test_split(totrain, test, test_size = 0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
reg = DecisionTreeRegressor()
reg.fit(X_train, y_train)
plt.figure(figsize = (10,7))
tree.plot_tree(reg, filled = True)
```

```python
path = reg.cost_complexity_pruning_path(X_train, y_train)
ccp_alpha = path.ccp_alphas
```

```python
alphalist = []
for i in range(len(ccp_alpha)):
  reg = DecisionTreeRegressor(ccp_alpha = ccp_alpha[i])
  reg.fit(X_train, y_train)
  alphalist.append(reg)
```

```python
trainscore = [alphalist[i].score(X_train, y_train) for i in range(len(alphalist))]
testscore = [alphalist[i].score(X_test, y_test) for i in range(len(alphalist))]

plt.xlabel('alpha')
plt.ylabel('accuracy')
plt.plot(ccp_alpha, trainscore, marker = 'o', label = 'training', color = 'magenta', drawstyle = 'steps-post')
plt.plot(ccp_alpha, testscore, marker = '+', label = 'testing', color = 'red', drawstyle = 'steps-post')
plt.legend()
plt.show()
```

```python
#Decision Tree Normal Approach
clf = DecisionTreeRegressor(ccp_alpha=1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(f"Decision Tree Normal Approach : {metrics.r2_score(y_test, y_pred)}")

# Random Forest Normal Approach 
rf = RandomForestRegressor(n_estimators = 1000, ccp_alpha =1)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print(f"Random Forest normal approach : {metrics.r2_score(y_test, y_pred_rf)}")
```

# Hyperparameter tuning

```python
params = {
    'RandomForest' : {
        'model' : RandomForestRegressor(),
        'params' : {
            'n_estimators' : [int(x) for x in np.linspace(start = 1, stop=1200, num = 10)],
            'max_depth' : [int(x) for x in np.linspace(start = 1, stop = 30, num = 5)],
            'min_samples_split' : [2,5,10,12],
            'min_samples_leaf' : [2,5,10,12],
            'max_features' : ['auto', 'sqrt'],
            'ccp_alpha' : [1,2,2.5,3,3.5,4,5],
        }
    },
    'D-tree' : {
        'model': DecisionTreeRegressor(),
        'params' : {
            'splitter' : ['best', 'random'],
            'min_samples_split' : [2,5,10,12],
            'min_samples_leaf' : [1,2,5,10,12],
            'max_features' : ['auto', 'sqrt'],
            'ccp_alpha' : [1,2,2.5,3,3.5,4,5],
        }
    },
    'SVM': {
        'model':SVR(),
        'params' : {
            'C':[0.25,0.50, 0.75, 1.0],
            'tol' : [1e-10, 1e-5, 1e-4, 0.025, 0.50, 0.75], 
            'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
            'max_iter' : [int(x) for x in np.linspace(start = 1, stop = 250, num = 10)],
        }
        
    }

}
```

```python
scores = []
for modelname, mp in params.items():
  clf = RandomizedSearchCV(mp['model'], 
                           param_distributions = mp['params'],
                           cv = 5, 
                           n_jobs = -1,
                           n_iter = 10, 
                           scoring = 'neg_mean_squared_error')
  clf.fit(X_train, y_train)
  scores.append({
      'model_name' : modelname, 
      'best_score' : clf.best_score_, 
      'best_estimator' : clf.best_estimator_
  })
```

```python
scoresdf = pd.DataFrame(scores, columns = ['model_name', 'best_score', 'best_estimator'])
scoresdf
```
